### 파이썬 클래스

```python
class Circle:
    pi = 3.14
    def __init__(self,radius):
        self.radius = radius
    def area(self):
        return self.radius ** 2 * Circle.pi

c = Circle(2.5)

print(c.area())
```

```python
class Rectangle:
    def __init__(self,widht,haight):
        self.widht = widht
        self.haight = haight
    def area(self):
        return self.widht * self.haight
    
rec = Rectangle(2.1,3.4)

print(rec.area())
```

```python
# 매직메서드
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def __str__(self):
        return f'{self.name}({self.age}살)'
    def __eq__(self,other):
        if self.name == other.name and self.age == other.age:
            return True
        return False
    
p1 = Person('홍길동',15)
print(p1)

p2 = Person('홍길동',20)
print(p2)

if p1 == p2:
    print('같은사람입니다')
else:
    print('다른사람입니다')
```

### 학습 내용 되돌아보기 (2)

```python
class Integer:
    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return self.value + other.value

    def __sub__(self, other):
        return self.value - other.value

a = Integer(5)
b = Integer(3)

print(a + b)
print(a - b)
```

### 학습 내용 되돌아보기 (4)

```python
class Student:
    def __init__(self, name, dept, mid, final):
        self.name = name
        self.dept = dept
        self.mid = mid
        self.final = final

    def __str__(self):
        return f"학과: {self.dept}, 이름: {self.name}, 중간: {self.mid}, 기말: {self.final}"

    def grade(self):
        avg = (self.mid + self.final) / 2
        if avg >= 90: return 'A'
        elif avg >= 80: return 'B'
        elif avg >= 70: return 'C'
        elif avg >= 60: return 'D'
        else: return 'F'

s1 = Student("김경철", "기계학과", 89, 90)
print(s1)
print(f"학점: {s1.grade()}")
```

### 학습 내용 되돌아보기 (5)

```python
class Car:
    def __init__(self, brand, year, color):
        self.brand = brand
        self.year = year
        self.color = color

    def __str__(self):
        return f"자동차 회사: {self.brand}, 년식: {self.year}, 색상: {self.color}"

    def __eq__(self, other):
        return (self.brand == other.brand and 
                self.year == other.year and 
                self.color == other.color)

mycar = Car('현대', 2020, '검정')
yourcar = Car('기아', 2021, '백색')

print(mycar)
print(yourcar)
print(mycar == yourcar)
```
