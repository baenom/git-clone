### 파이썬 클래스

```python
class Circle:
    pi = 3.14
    def __init__(self,radius):
        self.radius = radius
    def area(self):
        return self.radius ** 2 * Circle.pi

c = Circle(2.5)

print(c.area())
```

```python
class Rectangle:
    def __init__(self,widht,haight):
        self.widht = widht
        self.haight = haight
    def area(self):
        return self.widht * self.haight
    
rec = Rectangle(2.1,3.4)

print(rec.area())
```

```python
class Account:
    def __init__(self):
        self.__balance = 0
    def deposit(self, amount): 
        self.__balance += amount 
    def withdraw(self, amount):
        self.__balance -= amount
    def get_balance(self):
        return self.__balance

a1 = Account()
a1.deposit(100)
a2 = Account()
a2.deposit(50)

print("a1 : ", a1.get_balance())
print("a2 : ", a2.get_balance())
```

```python
# 매직메서드
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def __str__(self):
        return f'{self.name}({self.age}살)'
    def __eq__(self,other):
        if self.name == other.name and self.age == other.age:
            return True
        return False
    
p1 = Person('홍길동',15)
print(p1)

p2 = Person('홍길동',20)
print(p2)

if p1 == p2:
    print('같은사람입니다')
else:
    print('다른사람입니다')
```
