```python
class Book:
    def __init__(self, book_id, title, author, price, stock):
        self.book_id = book_id
        self.title = title
        self.author = author
        self.price = price
        self.stock = stock

    def __repr__(self):
        return f"[{self.book_id}] {self.title} - {self.author} ({self.price}원) | 재고: {self.stock}"
```

```python
class User:
    def __init__(self, user_id, username, password, address):
        self.user_id = user_id
        self.username = username
        self.password = password
        self.address = address  
        self.cart = []   

    def add_to_cart(self, book, quantity=1):
        if book.stock >= quantity:
            self.cart.append({"book": book, "quantity": quantity})
            print(f"'{book.title}' {quantity}권이 장바구니에 담겼습니다.")
        else:
            print(f"재고가 부족합니다. (현재 재고: {book.stock})")

    def checkout(self):
        if not self.cart:
            print("장바구니가 비어 있습니다.")
            return

        total_price = sum(item['book'].price * item['quantity'] for item in self.cart)
        
        print("주문서")
        print(f"수령인: {self.username}")
        print(f"배송지: {self.address}")
        for item in self.cart:
            book = item['book']
            qty = item['quantity']
            book.stock -= qty # 재고 차감
            print(f"- {book.title} x {qty}")
        
        print(f"총 결제 금액: {total_price}원")
        print("배송을 시작합니다! [cite: 15, 40]")
        self.cart = []
```

```python
class Admin:
    def __init__(self, admin_password):
        self.admin_password = admin_password

    def add_book(self, library, book):
        library.append(book)
        print(f"관리자: 새 도서 '{book.title}'가 추가되었습니다.")

    def delete_book(self, library, book_id):
        for i, book in enumerate(library):
            if book.book_id == book_id:
                removed = library.pop(i)
                print(f"관리자: 도서 '{removed.title}'가 삭제되었습니다.")
                return
        print("해당 ID의 도서를 찾을 수 없습니다.")
```
