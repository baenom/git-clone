```python
# 구구단
i = int(input())
def gugudan(num):
    print(f'[구구단 {num}단 ]')
    for i in range(1,10):
        print(f'{num} x {i} = {num*i}')

gugudan(i)
```

```python
for i in range(2,10):
    gugudan(i)
    print()
```

```python
def add_and_mul(a,b):
    return a+b, a*b

result = add_and_mul(3,4)
print(result)

result1, result2 = add_and_mul(3,4)
print(result1, result2)
```
