### 문자열을 원하는 형식으로 만들기 : f-문자열 (Formatted String)

```python
print (f"2 + 3 = (2 + 3)")
lecture = 'python'
print (f"Hello, {lecture. upper ()}!")
n = 10
print(f"{n}은 {'짝수' if n % 2 == 0 else '홀수'}입니다.")
```

```python
number = 1234567.89123
print(f"소수점 자리수 제한: {number: 2f}")
print(f"콤마 추가: {number:,}")
print(f" 동시 사용: {number:,.2f}")
```

### 문자열을 원하는 형식으로 만들기

```python
string = "ABC"
print (f"| {string:10}|") 
print (f"| {string:>10}|") 
print (f"| {string:-^10}|") 
print (f"| {string:*>10}|")
```

```python
number = 1234
print (f"| {number:10}|")
print (f"| {number:<10} |") 
print (f"| {number:-^10} |") 
print (f"| {number:*>10}|")
print(f"|{number:0>10}|")
print(f"|{number:010}|")
```

### [실습] 모스 부호 번역기

```python
morse = (".-", "-...", "-.-.", "-..", ".", "..-.", "--.", "....", "..", ".---", "-.-", ".-..", "--", "-.", "---", ".--.", "--.-", ".-.", "...", "-", "..-", "...-", ".--", "-..-", "-.--", "--..")

strin = input()

def mo(i):
    ch = int(ord(i))
    return morse[ch - 65]


for i in strin:
    print(mo(i),end=" ")
```
