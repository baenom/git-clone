```python
s1 = set ([1, 2, 3, 4, 5, 6])

s2 = set ([4, 5, 6, 7, 8, 91])
s_inter1 = s1 & s2
print(s_inter1)

s_inter2 = s1. intersection(s2)
print(s_inter2)
s_union1 = s1 | s2
print(s_union1)

s_union2 = s1.union(s2)
print(s_union2)
s_diff1 = s1 - s2
print(s_diff1)


s_diff2 = s1.difference (s2)
print(s_diff2)
s_symdiff1 = s1^s2
print(s_symdiff1)

s_symdiff2 = s1. symmetric_difference(s2)
print(s_symdiff2)
```

```python
S = input().strip()

resurt = set()

for i in range(len(S)):
    for j in range(i,len(S)):
        s = S[i:j+1]
        resurt.add(str(s))

print(len(resurt))
```

```python
import sys

input = sys.stdin.readlines

num = int(input())

person = set()
for i in range(num):
    name, action = input().split()
    if action == 'enter':
        person.add(name)
    else:
        person.remove(name)

print('\n'.join(sorted(person, reverse=True)))
```
