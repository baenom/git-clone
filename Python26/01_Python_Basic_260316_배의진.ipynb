### 02. 대입연산자
* 두변수의 값을 교환하는 프로그램(swap)

```python
a,b = 10,20
print(a,b)
temp = a
a = b
b = temp
print(a,b)
```

```python
a = b = c = 7
print(a,b,c)
```

```python
a,b,c = 7,11,19
print(a,b,c)
```

```python
a,b = 10,20
print(a,b)
a,b = b,a
print(a,b)
```

```python
a = 10>20
print(a)
```

### 03. 나의이름 출력하기

```python
name_1 = "baeeuijin"
print(name_1)
name_2 = "인공지능 소프트웨어과"
print(name_1)
name_3 = '인공지능 소프트웨어과 "baeeuijin"'
print(name_3)
name_4 = "인공지능 소프트웨어과 'baeeuijin'"
print(name_4)
name_5 = '''인공지능 소프트웨어과 
"baeeuijin"'''
print(name_5)
```

### 06. 원의 넓이 구하기

```python
pi = 3.14159265
radius = int(input("반지름:"))
area = radius * radius * pi
print(f"원의 반지름이 {radius}이라면 면적은 {area:.3f}입니다")
```
