### 튜플 (tuple)의 개념

```python
t1 = (1, 2, 3)

t2 = ('a','b', ('ab', 'cd'))


t3 = ('tuple', (1, 2), [3, 4])
print(t1)
print(t2)
print(t3)
```

### 튜플(tuple) 만들기

```python
t4 = ()
t5 = (1,)
print(t4)
t6 = 1, 2, 3
print(t5)
t7 = 1,
print(t7)
print(t6)
x, y, z = t6
print (x, y, z)
```

### 튜플의 활용

```python
x, y = 10, 20
x, y = y, x
print(x,y)
```
