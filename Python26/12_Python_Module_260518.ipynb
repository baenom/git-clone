### 학습 내용 되돌아보기(2)

```python
import math
import random


def getarea(r):
    return math.pi * (r ** 2)

radius = round(random.random() * 10, 2)

area = getarea(radius)

print(f"원의 반지름: {radius}")
print(f"원주율 pi: {round(math.pi, 4)}")
print(f"반지름 {radius}인 원의 면적은 {round(area, 2)}")
```

```python
import math
import statistics

if __name__ == '__main__':
    for n in range(1, 17, 5):
        print(f"{n}! = {math.factorial(n)}")
        
    print()

    st = [80, 99, 77, 65, 92, 74, 82]
    print(st)
    print(f"중앙 값: {statistics.median(st):.2f}")
    print(f"평균: {statistics.mean(st):.2f}")
    print(f"분산: {statistics.variance(st):.2f}")
    print(f"표준편차: {statistics.stdev(st):.2f}")
```
