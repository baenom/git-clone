```python
while True:
    try:
        number = int(input('양의정수'))
        if number <= 0:
            raise ValueError()
    except ValueError:
        print('양의정수를 입력하세요')
    else:
        sum = 0
        for i in range(1,number+1):
            sum += i
        print(f'1부터 {number}까지의 합 = {sum}')
        break
```

```python
def divide(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        print("0으로는 나눌 수 없습니다.")
    else:
        print(f"결과: {result}")

divide(3.2, 2)
divide(5.4, 0)
```

```python
uyear = {1: 'freshman', 2: 'sophomore', 3: 'junior', 4: 'senior'}

try:
    m = int(input('대학교 몇 학년이지요? '))
    
    if not (1 <= m <= 4):
        raise Exception('1~4 정수를 입력하세요.')

except Exception as e:
    print(f"예외 발생 이름: {type(e)}")
    print(f"예외 발생 이유: {e}")

else:
    print(f"{m}학년: {uyear[m]}")

finally:
    print("예외 처리가 잘되는군요!")
```
