```python
while True:
    try:
        number = int(input('양의정수'))
        if number <= 0:
            raise ValueError()
    except ValueError:
        print('양의정수를 입력하세요')
    else:
        sum = 0
        for i in range(1,number+1):
            sum += i
        print(f'1부터 {number}까지의 합 = {sum}')
        break
```
