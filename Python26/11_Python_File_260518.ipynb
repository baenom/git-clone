```python
filename = './sample.txt'

with open(filename, encoding='utf-8') as f_read:
    content = f_read.read()
print(content)
```

```python
with open(filename,encoding='utf-8') as f_read:
    lines = f_read.readlines()

for line in lines:
    print(line.strip())
```

```python
with open(filename,encoding='utf-8') as f_read:
    while True:
        lines = f_read.readline()
        if lines: print(lines.strip())
        else: break
```

```python
fruits = ['사과','토마토','바나나']
outfile = 'fruits.txt'
with open(outfile,mode = 'w',encoding='utf-8') as f_write:
    for fruit in fruits:
        f_write.write(fruit+'\n')

with open(outfile, mode='a',encoding='utf-8') as f_write:
    f_write.write('\n'.join(fruits))
```

```python
with open(filename, mode='rb') as f_read:
    content = f_read.read()
print(content)
print(content.decode('utf-8'))
```

```python
# 학습 내용 되돌아보기(1)

lst = ['text file\n', 'binary file\n']

with open('check.txt', 'w') as f:
    f.writelines(lst)

with open('check.txt') as f:
    print(f.read())
```

```python
# 학습 내용 되돌아보기 (2)

f = open('pyzenex.txt')
print(f.read())
f.close()

f = open('pyzenex.txt')
for line in f.readlines():
    print(line, end='')
f.close()

f = open('pyzenex.txt')
for line in f:
    print(line, end='')
f.close()

f = open('pyzenex.txt')
while True:
    line = f.readline()
    if not line: break
    print(line, end='')
f.close()
```

```python
# 학습 내용 되돌아보기 (3)

with open('cp.txt', 'w') as f:
    f.write('freshman\n')
    f.write('sophomore\n')
    f.write('junior\n')
    f.write('senior\n')

with open('cp.txt', 'r') as f:
    for i, line in enumerate(f, start=1):
        print(f"{i}학년 {line.strip()}")
```
