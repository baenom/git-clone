```python
# 3개의 숫자중 최대값 구하기
#a,b,c = input("3개의 숫자 입력").split( )
a,b,c = map(int, input("3개의 숫자 입력").split( ))
# a = int(a)
# b = int(b)
# c = int(c)

max = a

if b > max:
    max = b
if c > max:
    max = c
print(f"가장 큰수는 {max}입니다")
```

```python
age1 = int(input("첫번째 사람의 나이:"))
age2 = int(input("두번째 사람의 나이:"))

if age1 > age2:
    diff = age1 - age2
else:
    diff = age2 - age1

print("나이차이 => ",diff)
```

```python
# 입력된 점수에 따라 등급 출력
score = int(input("점수를 입력하세요:"))


if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f">> 등급은 {grade}입니다")
```

```python
#백준 연습
N,M,K = map(int, input().split())

if K >= M:
    n = K // M
    m = K % M
else:
    n = 0
    m = K % M
print(n,m)

```

```python
#놀이공원 기본 입장료가 5,000원이고 나이에 따른 할인율이 다음과 같은 경우, 나이를 입력받아 입장료를
#구하기
age = int(input("나이를 입력하세요:"))
pay = 5000
if age < 8:
    pay = pay * (1 - 0.30)
elif age >= 60:
    pay = pay * (1 - 0.20)
print(f"{age}살의 입장료는 {pay}원입니다")

```

```python
# while 문 연습

tree_hit = 0
while tree_hit < 10:
    tree_hit += 1
    print(f"나무를 {tree_hit}번 찍었습니다")

print("나무가 넘어갑니다")
```

```python
# while 문 연습

num = int(input("숫자 입력 (0 : 끝)"))
sum = count = 0
while num != 0:
    sum += num
    count += 1
    num = int(input("숫자 입력 (0 : 끝)"))


if count != 0:
    average = sum / count
else:
    average = 0



print(f"총합 : {sum}, 평균 : {average:.2f}")
```

```python
# while 문 무한 반복 연습

sum = count = 0
while True:
    
    num = int(input("숫자 입력 (0 : 끝)"))
    if num == 0:
        break
    sum += num
    count += 1

if count != 0:
    average = sum / count
else:
    average = 0



print(f"총합 : {sum}, 평균 : {average:.2f}")
```

```python
#for문 연습

sum = 0
for i in 1.1,2.5,3.6,4.2,5.4:
    sum += i
    print(i,sum)

print(f"합 : {sum}, 평균 : {sum / 5:.2f}")
```

```python

number = int(input("숫자를 입력하세요 :"))

sum = 0
for num in range(1, number + 1):
    sum += num

print(f"1부터 {number}까지의 값은 {sum}입니다.")
```

```python
num = int(input("숫자를 입력하세요 : "))

facto = 1

for i in range(num,0,-1):
    facto *= i

print(f"{num}! = {facto}")
```
