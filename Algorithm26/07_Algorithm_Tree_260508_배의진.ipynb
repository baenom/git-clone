```python
d = TreeNode('D',None,None)
e = TreeNode('E',None,None)
b = TreeNode('B',d,e)
f = TreeNode('F',None,None)
c = TreeNode('C',f,None)
a = TreeNode('A',b,c)
my_bynary_tree = a
```

```python
# preorder (선위순회)
# a,b,d,e,c,f

def preorder(node):
    if node is None:
        return
    print(node.data,end='')
    if node.left: preorder(node.left)
    if node.right: preorder(node.right)

preorder(my_bynary_tree)
```

```python
# inorder (중위 순회)
# d,b,e,a,f,c

def inorder(node):
    if node is None:
        return

    inorder(node.left)
    print(node.data, end=' ')
    inorder(node.right)

inorder(my_bynary_tree)
```

```python
# postorder (후위 순회)
# d,e,b,f,c,a

def postorder(node):
    if node is None:
        return
    if node.left: postorder(node.left)
    if node.right: postorder(node.right)
    print(node.data, end=' ')

postorder(my_bynary_tree)
```

```python
class STreeNode:
    def __init__(self,key,left,right):
        self.key = key
        self.left = left
        self.right = right
```

```python
t3 = STreeNode(3,None,None)
t12 = STreeNode(12,None,None)
t7 = STreeNode(7,t3,t12)
t27 = STreeNode(27,None,None)
t31 = STreeNode(31,t27,None)
t26 = STreeNode(26,None,t31)
t18 = STreeNode(18,t7,t26)
my_bst = t18
```

```python
def search_bst(n,key):
    if n == None: return None
    if key == n.key: return n
    if key < n.key:
        return search_bst(n.left,key)
    else:
        return search_bst(n.right,key)

result = search_bst(my_bst,12)
if result: print(result.key)
else:
    print('none')
```

```python
def search_bst_iter(n,key):
    while n != None:
        if key == n.key: return n
        elif key < n.key:
            n = n.left
        else:
            n = n.right
    return None

result = search_bst_iter(my_bst,12)
if result: print(result.key)
else:
    print('none')
```
