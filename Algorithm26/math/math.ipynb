```python
# 최대 공약수

def getGCD(M,N):
    if M > N: M,N = N,M

    for num in range(M,0,-1):
        if M % num == 0 and N % num == 0:
            return num
        
print(getGCD(12,18))
```

```python
# 최소 공배수

def getLCM(M,N):
    if M > N: M,N = N,M

    i = 1
    while True:
        multiple = N * i
        if multiple % M == 0:
            return multiple
        i += 1
        
print(getLCM(12,18))
```

```python
num = int(input())
Num = int(input())

def two(num,Num):
    out_list = ''
    digits = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    while True:
        out,num = num % Num ,num // Num
        out_list = str(out) + out_list if out < 10 else digits[out] + out_list
        if num == 0: break
    return out_list

print(two(num,Num))
```
