```python
# 최대 공약수

def getGCD(M,N):
    if M > N: M,N = N,M

    for num in range(M,0,-1):
        if M % num == 0 and N % num == 0:
            return num
        
print(getGCD(12,18))
```

```python
# 최소 공배수

def getLCM(M,N):
    if M > N: M,N = N,M

    i = 1
    while True:
        multiple = N * i
        if multiple % M == 0:
            return multiple
        i += 1
        
print(getLCM(12,18))
```

### 투포인터

```python
N,M = map(int,input().split())
A = list(map(int,input().split()))
B = list(map(int,input().split()))
new_list = []

pA = 0
pB = 0
while pA < N and pB < M:
    if A[pA] < B[pB]:
        new_list.append(A[pA])
        pA += 1
    else:
        new_list.append(B[pB])
        pB += 1

while pA < N:
    new_list.append(A[pA])
    pA += 1

while pB < M:
    new_list.append(B[pB])
    pB += 1

for i in new_list:
    print(i, end=" ")
```

```python
N,M = map(int,input().split())
num_list = list(map(int,input().split()))
count = 0
start = end = 0
sum = num_list[0]
while end < N:
    if sum == M: count += 1
    if sum <= M:
        end += 1
        if end < N: 
            sum += num_list[end]
    else:
        sum -= num_list[start]
        start += 1
print(count)
```

### 진법 변환

```python
#진법 변환기

num = int(input())
Num = int(input())

def two(num,Num):
    out_list = ''
    digits = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    while True:
        out,num = num % Num ,num // Num
        out_list = str(out) + out_list if out < 10 else digits[out] + out_list
        if num == 0: break
    return out_list

print(two(num,Num))
```
