```python
d = TreeNode('D',None,None)
e = TreeNode('E',None,None)
b = TreeNode('B',d,e)
f = TreeNode('F',None,None)
c = TreeNode('C',f,None)
a = TreeNode('A',b,c)
my_bynary_tree = a
```

```python
# preorder (선위순회)
# a,b,d,e,c,f

def preorder(node):
    if node is None:
        return
    print(node.data,end='')
    if node.left: preorder(node.left)
    if node.right: preorder(node.right)

preorder(my_bynary_tree)
```

```python
# inorder (중위 순회)
# d,b,e,a,f,c

def inorder(node):
    if node is None:
        return

    inorder(node.left)
    print(node.data, end=' ')
    inorder(node.right)

inorder(my_bynary_tree)
```

```python
# postorder (후위 순회)
# d,e,b,f,c,a

def postorder(node):
    if node is None:
        return
    if node.left: postorder(node.left)
    if node.right: postorder(node.right)
    print(node.data, end=' ')

postorder(my_bynary_tree)
```

```python
tree = {}
tree['A'] = ['B','C']
tree['B'] = ['D','E']
tree['C'] = ['F',None]
tree['D'] = [None,None]
tree['E'] = [None,None]
tree['F'] = [None,None]
```

```python
def preorder(tree, node):
    if node is None:
        return

    print(node, end=' ')
    preorder(tree, tree[node][0])
    preorder(tree, tree[node][1])

preorder(tree, 'A')
```

```python
def inorder(tree, node):
    if node is None:
        return

    inorder(tree, tree[node][0])
    print(node, end=' ')
    inorder(tree, tree[node][1])

inorder(tree,'A')
```

```python
def postorder(tree, node):
    if node is None:
        return
    if tree[node][0]: postorder(tree, tree[node][0])
    if tree[node][1]: postorder(tree, tree[node][1])
    print(node, end=' ')

postorder(tree,'A')
```
