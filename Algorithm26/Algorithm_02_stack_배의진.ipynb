```python
stack = []

while True:
    print('============================')
    print('1.Push 2.Pop 3.Peak 0.Exit')
    print('============================')
    menu = int(input('>> 연산 메뉴 선택 :'))
    if menu == 0:
        break
    elif menu == 1: 
        data = int(input('[Push] 추가할 데이터 입력 :'))
        stack.append(data)
    elif menu == 2: 
        data = stack.pop()
        print('[Pop] 가져온 데이터 :', data)
    elif menu == 3:
        data = stack[len(stack)-1]
        print('top 데이터 확인:', data)
        continue

    print('[Stack]', stack)

print('정수형 스택 종료')
```

```python
# Boj 10828
import sys
stack = []

N = int(sys.stdin.readline())

for _ in range(N):
    operation = sys.stdin.readline().split()
    if operation[0] == 'push':
        stack.append(int(operation[1]))
    elif operation[0] == 'pop':
        print(stack.pop()) if stack else print(-1)
    elif operation[0] == 'size':
        print(len(stack))
    elif operation[0] == 'empty':
        print(1 if not stack else 0)
    elif operation[0] == 'top':
        print(stack[-1] if stack else -1)
```

```python
while True:

    char = input()
    if char == '.':
        break
    out = True
    stack = []
    if char[0] == ')' or char[0] == ']':
        out = False
    else:
        for i in char:
            if i == '(' or i == '[':
                stack.append(i)
            elif i == ')':
                if len(stack) != 0 and stack[-1] == '(':
                    stack.pop()
                else:
                    out = False
                    break
            elif i == ']':
                if len(stack) != 0 and stack[-1] == '[':
                    stack.pop()
                else:
                    out = False
                    break
            
            if i == '.':
                break
    if out and not stack:
        print('yes')
    else:
        print('no')
```
