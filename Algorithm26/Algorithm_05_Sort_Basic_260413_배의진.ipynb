### 선택 정렬 (Selection Sort)

```python
A = [3,5,2,6,7,8,9,1]

def select(A,start):
    min = A[start]
    min_index = start
    for i in range(start+1,len(A)):
        if A[i] < min:
            min = A[i]
            min_index = i
    return min_index

def print_step(step,A):
    print(f"{step+1:2}단계 : {A}")

def sort_list(data_list):
    for start in range(len(data_list)-1):
        min_index = select(data_list,start)
        data_list[start], data_list[min_index] = data_list[min_index], data_list[start]
        print_step(start,data_list,)
print(f"정렬전{A}")
sort_list(A)
print(f"정렬후{A}")
```

### 삽입 정렬 (Insertion Sort)

```python
def print_step(step,A):
    print(f"{step:2}단계 : {A}")

def insertion_sort_list(A):
    for cur_index in range(1,len(A)):
        key = A[cur_index]
        i = cur_index -1
        while i >= 0 and A[i] >= key:
            A[i+1] = A[i]
            i -= 1
        A[i+1] = key
        print_step(cur_index,A)
```

```python
A = [3,5,50,41,30,8,9,1]      
print(f"정렬전{A}")
insertion_sort_list(A)
print(f"정렬후{A}")
```

### 버블 정렬 (Bubble Sort)

```python
def bubble_sort_list(A):
    for last in range(len(A)-1,0,-1):
        is_changed = False
        for i in range(last):
            if A[i] > A[i+1]:
                A[i], A[i+1] = A[i+1],A[i]
                is_changed = True
        print_step(last,A)
        if not is_changed: break 
    return A

def print_step(step,A):
    print(f"{step:2}단계 : {A}")

A = [23,5,44,1,6,45,46]      
print(f"정렬전{A}")
bubble_sort_list(A)
print(f"정렬후{A}")
```
