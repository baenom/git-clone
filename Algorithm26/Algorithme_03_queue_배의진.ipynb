```python
from collections import deque
queue = deque()


def enqueue(data):
    queue.append(data)

print("시작")

def dequeue():
    if queue:
        data = queue.popleft()
    else:
        data = -1
    return data

def getfront():
    if queue:
        return queue[0]
    else:
        return -1
    
while True:
    print('======================================')
    print('1.Enqueue 2.Dequeue 3.Get Front 0.Exit')
    print('======================================')
    menu = int(input(">>"))

    if menu == 0: break
    elif menu == 1:
        data = int(input("데이터 : "))
        enqueue(data)
    elif menu == 2:
        data = dequeue()
        if data != -1:
            print("가져온 데이터 : ", data)
        else:
            print('Queue가 비어있습니다')
    elif menu == 3:
        data = getfront()
        if data != -1:
            print(data)
        else: 
            print('Queue가 비어있습니다')

        
print("종료")
```
