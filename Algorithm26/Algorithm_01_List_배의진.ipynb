```python
# 파이썬 리스트 연습

test_list = [1,2,3,'one','two','three']
print(test_list[0])
print(test_list[3])
print(test_list[-1])
print(test_list[-3])

for item in test_list:
    print(item)
```

```python
#파이썬 리스트 연습 : 항목 삽입과 삭제
test_list = [1,2,3,'one','two','three']


test_list.insert(3,4)
test_list.append('four')
test_list.remove('one')
item = test_list.pop(0)
print(item)
item = test_list.pop()

print(item)
print(test_list)
```

```python
#리스트와 반복구조

data_list = [23, 45, 21, 14, 2, 67]
target = 14
index = 0
while index <= 5:
    if data_list[index] == target:
        break

    index += 1
print(index)
```

```python
#리스트와 반복구조

data_list = [23, 45, 21, 14, 2, 67]
target = 14
for index in range(6):
    if data_list[index] == target:
        break

print(index)
```

```python
data_list = [123, 45, 10, 14, 2, 67]

max = data_list[0]
max_index = 0
for i in range(1,6):
    if max < data_list[i]:
        max = data_list[i]
        max_index = i

print(max, max_index)
```

```python
data_list = [123, 45, 10, 14, 2, 67]

max = data_list[0]
for i in data_list:
    if i > max:
        max = i

print(max)
```

```python
#최솟 값
data_list = [123, 45, 10, 14, 2, 67]

min = data_list[0]
for i in data_list:
    if i < min:
        min = i

print(min)
```
