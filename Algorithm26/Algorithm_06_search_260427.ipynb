```python
def binary_search_recursive(A,key,low,high):
    if low > high:# 만약 low가 high보다 크면 위치가 뒤바뀌었으므로 다 돌려보았을때 key가 존재하지 않는다는 이야기이므로 -1을 리턴한다
        return -1
    
    middle = (low+high)//2 # 미들의 값을 업데이트 하기 위해 현재 상태를 판별하기 전 low와 high를 더한 값에서 2로 나눈값 즉 중앙값을 찾아 업데이트 한다

    if key == A[middle]: #만약 key와 A의 middle번째 인덱스가 같다면 원하는 값을 찾았다는 뜻이므로 미들값을 리턴하며 함수를 종료한다
        return middle
    elif key < A[middle]: #만약 key가 A의 middle번째 인덱스 보다 작다면 A리스트의 미들의 오른쪽에는 값이 존재하지 않다는 명백한 뜻이므로 왼쪽의 값만 다시 찾기위해 하이 자리를 미들-1의 위치로 설정후 본인의 함수를 다시 실행한다
        return binary_search_recursive(A,key,low,middle-1)
    else: #만약 위의 두 경우 다 아니라면 key가 A의 middle번째 인덱스 보다 크다는 증거이다 A리스트의 미들의 왼쪽에는 값이 존재하지 않다는 명백한 뜻이므로 오른쪽의 값만 다시 찾기위해 로우 자리를 미들+₩1의 위치로 설정후 본인의 함수를 다시 실행한다
        return binary_search_recursive(A,key,middle+1,high)
```

```python
def binary_search_loop(A,key):
    low = 0 # 로우의 초기 값을 정해주기위해 모든 리스트의 첫번째인 인덱스 0을 집어넣었다
    high = len(A) - 1 # 하이를 설정하기위해 리스트의 전체길이에서 1을 뺀값을 하이에 대입시킨다
    while low <= high: # 하이와 로우의 위치가 뒤집어져있지않다면 계속 실행한다
        middle = (low + high) //2 #판별전 미들의 값을 업데이트 하기위해 중앙값을 미들에 업데이트한다
        if A[middle] == key: #만약 key와 A의 middle번째 인덱스가 같다면 원하는 값을 찾았다는 뜻이므로 미들값을 리턴하며 함수를 종료한다
            return middle 
        elif A[middle] > key: # #만약 key가 A의 middle번째 인덱스 보다 작다면 A리스트의 미들의 오른쪽에는 값이 존재하지 않다는 명백한 뜻이므로 왼쪽의 값만 다시 찾기위해 하이에 미들-1을 대입한다
            high = middle - 1
        else:
            low = middle + 1# #만약 key가 A의 middle번째 인덱스 보다 크다면 A리스트의 미들의 왼쪽에는 값이 존재하지 않다는 명백한 뜻이므로 오른쪽의 값만 다시 찾기위해 로우에 미들+1을 대입한다
    return -1 #여기까지 왔다면 하이와 로우의 위치가 뒤집어졌다는 뜻이므로 찾는 수가 존재하지 않는다는 의미로 -1을 리턴한다
```

```python
test_list = [1,2,3,5,6,7,9]

num = int(input())

print('찾을 숫자',num)

index = binary_search_loop(test_list,num)

if index == -1:
    print('여기에는 찾는 수가 없습니다')
else: print(f"루프문으로 찾은 수는 {index}번째 인덱스에 있습니다")
index = binary_search_recursive(test_list,num,0,6)
if index == -1:
    print('여기에는 찾는 수가 없습니다')
else: print(f"재귀함수로 찾은 수는 {index}번째 인덱스에 있습니다")
```

```python
test_list = [1,2,3,5,6,7,9]

answer = [10,0,4,5,6,7,2]

for i in answer:
    index = binary_search_loop(test_list,i)
    if index == -1:
        print('1.루프문으로 찾았는데 여기에는 찾는 수가 없습니다')
    else: print(f"1.루프문으로 찾은 수는 {index}번째 인덱스에 있습니다")
    index = binary_search_recursive(test_list,i,0,6)
    if index == -1:
        print('2.재귀함수로 찾았는데 여기에는 찾는 수가 없습니다')
    else: print(f"2.재귀함수로 찾은 수는 {index}번째 인덱스에 있습니다")
```
